```python
with open('/content/input.txt', 'r') as f:
    calibration = f.read().split('\n')

# Replace the number digit "1" with "one1one" and then reduce to part "1"

clues = {"one": "one1one", "two": "two2two", "three": "three3three", "four": "four4four", "five": "five5five", "six": "six6six", "seven": "seven7seven", "eight": "eight8eight", "nine": "nine9nine"}

def decrypter(calibration, clues):

    progression = calibration[:]

    for i, calibration in enumerate(progression):
        for key in clues:
            progression[i] = progression[i].replace(key, clues.get(key))

    return progression

new_values = decrypter(calibration, clues)
print(new_values)
```

```python
first_last = []
lista_int = []
total = 0

#Delete letters and add to a new list "no_letters"
def delete_letters(new_values):
  no_letters = []
  for i in new_values:
    if i.isnumeric():
      no_letters.append(i)

#Add via .append() to the list "no_letters" only the 1st and last number and added to the list "first_last"
  first_last.append(no_letters[0]+no_letters[-1])
for i in new_values:
  delete_letters(i)
for j in range(len(first_last)):

  x = float(first_last[j])

  lista_int.append(x)
total = sum(lista_int)
print(f"The sum of calibration values is: ", total)
```
